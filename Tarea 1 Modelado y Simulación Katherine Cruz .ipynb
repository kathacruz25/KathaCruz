$1.$ Demostrar que solamente existe 5 sólidos regulares.

## Solución.

**Definición (poliedro):** Los cuerpos geométricos existen en el espacio y son, por tanto, objetos tridimensionales limitados por una o varias superficies. Si todas las superficies que lo limitan son planas y de contorno poligonal, el cuerpo es un poliedro.



**Demostración:**
Queremos ver que solo hay cinco sólidos regulares. Tenemos la formula de Euler: 
$$C - A + V = 2 \quad \quad(1)$$

Entonces, Sea C el número de caras de un poliedro, si contamos todos los lados de todas las caras de un poliedro, es decir nC, podemos ver que pasa dos veces por las aristas, de manera, que:
$$nC = 2A \quad\quad(2)$$.

Sea V el número de vértices, entonces si contamos todos los vértices que cada arista conecta, es decir sV, tenemos:
$$sV = 2A \quad\quad(3)$$

Ahora, si despejamos (2) en términos de $C$, obtenemos:

\begin{equation}
\begin{split}
nC &
= 2A \\&
= \frac{2A}{n}
\end{split}
\end{equation}

Si despejamos (3) en términos de $V$, obtenemos:

\begin{equation}
\begin{split}
sV &
= 2A \\&
= \frac{2A}{s}
\end{split}
\end{equation}

Si sustituimos los valores de $C$ y $V$ en (1), tenemos:

\begin{equation}
\begin{split}
\frac{2A}{n} - A + \frac{2A}{s} = 2 \quad \quad (4)
\end{split}
\end{equation}

Entonces

\begin{equation}
\frac{2A}{2A(n)} - \frac{A}{2A} + \frac{2A}{2A(s)} = \frac{2}{2A}\\
\end{equation}

\begin{equation}
\Rightarrow \frac{1}{n} - \frac{1}{2} + \frac{1}{r} = \frac{1}{A}
\end{equation}

\begin{equation}
\Rightarrow \frac{1}{n} + \frac{1}{s} = \frac{1}{A} + \frac{1}{2}
\end{equation}

Tenemos que $n \geq 3$ y $s \geq 3$ porque dado un vértice de un poliedro se encuentran por lo menos $3$.

* Si $n = 3$, entonces:

\begin{equation}
\frac{1}{3} + \frac{1}{s} = \frac{1}{A} + \frac{1}{2}
\end{equation}

Operando obtenemos:

\begin{equation}
\frac{1}{s} = \frac{1}{A} + \frac{1}{6}
\end{equation}

Con $s = 3,4,5$. Para que satisfaga la ecuación anterior.Para que satisfaga la ecuación anterior. Entonces:
* Si $r = 4$ forma el octaedro.
* Si $r = 5$ forma el icosaedro.

Por otro lado
* Si $s = 3$

\begin{equation}
\frac{1}{n} + \frac{1}{3} = \frac{1}{A} + \frac{1}{2}
\end{equation}

Operando obtenemos:

\begin{equation}
\frac{1}{n} = \frac{1}{A} + \frac{1}{6}
\end{equation}

Con $s = 3,4,5$. Para que satisfaga la ecuación anterior. Entonces:
* Si $n = 3$ es el tetraedro.
* Si $n = 4$ es el cubo.
* Si $n = 5$ es el dodecaedro.

Así, tenemos que solo hay cinco sólidos regulares, puesto que no tenemos más valores enteros positivos para $n$ y $r$. $\blacksquare$
